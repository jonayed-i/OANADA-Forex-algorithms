```python
stocks = pd.read_csv("stocks1.csv",header=[0,1],index_col=[0],parse_dates=[0])
```

```python
import matplotlib.pyplot as plt
close = stocks.loc[:,"Adj Close"].copy()
close.plot(figsize = (15,8), fontsize = 13)
plt.style.use("seaborn")
plt.legend(fontsize = 13)
plt.show()



```

```python
close = close.div(close.iloc[0]).mul(100)

```

```python
close.plot(figsize = (15,8), fontsize = 13)
plt.style.use("seaborn")
plt.legend(fontsize = 13)
plt.show()
```

```python
nvda = close.NVDA.copy().to_frame()
```

```python
nvda["nextday"]=nvda.shift(periods = 1)
```

```python
nvda["change"] = nvda.NVDA.sub(nvda.nextday)
nvda.head(10)
```

```python
nvda["relchange"] = nvda.NVDA.div(nvda.nextday).sub(1).mul(100)
nvda.head(10)
```

```python
nvda.relchange.plot(figsize = (15,8), fontsize = 13)
plt.style.use("seaborn")
plt.legend(fontsize = 13)
plt.show()
```

```python
nvda["pctdif"] = nvda.NVDA.pct_change(periods = 1).mul(100).dropna()
nvda.tail(10)
```

```python
ret = nvda.pctdif.dropna()
ret.head(5)
ret.plot(figsize = (15,8), fontsize = 13)
plt.style.use("seaborn")
plt.legend(fontsize = 13)
plt.show()
```

```python
ret.plot(kind="hist",figsize = (12,8),bins=100)
ret.describe()
```
